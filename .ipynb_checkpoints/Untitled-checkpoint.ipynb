```python
import time
from tqdm.auto import tqdm

def label_row_with_gpt4(title, abstract):
    prompt = f"""You are a quantum computing expert.

Below is the TITLE and ABSTRACT of a research paper.  
Your task: choose **exactly one** label from this list.  
If you are not reasonably certain which category fits best, respond **exactly** with "unknown".

Labels:
{', '.join(labels)}

Title:
\"\"\"{title}\"\"\"

Abstract:
\"\"\"{abstract}\"\"\"

Label:"""

    resp = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system",  "content": "You are a helpful assistant."},
            {"role": "user",    "content": prompt}
        ],
        temperature=0
    )
    return resp.choices[0].message.content.strip()



data['Label'] = None
data['Error'] = None


for idx, row in tqdm(data.iterrows(), total=len(data), desc="Labeling abstracts"):
    try:
        title    = row['Title']
        abstract = row['Abstract']
        data.at[idx, 'Label'] = label_row_with_gpt4(title, abstract)
    except OpenAIError as e:
        # record the error message
        data.at[idx, 'Error'] = str(e)
        data.at[idx, 'Label'] = "error"
    time.sleep(1)  
```
